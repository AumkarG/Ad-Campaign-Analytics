### Using Inter Quartile Range

```python
Q1=m.quantile(0.25)
Q3=m.quantile(0.75)
print(Q1)
print(Q3)
```

```python
IQR=Q3-Q1
print(IQR)
```

```python
m1=m[(m < (Q1 - 1.5 * IQR)) |(m > (Q3 + 1.5 * IQR))]
m1
```

```python
m1.dropna(axis = 0, how = 'all', inplace = True)
```
